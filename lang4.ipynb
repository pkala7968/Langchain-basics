```python
import pandas as pd

pd.set_option('display.max_colwidth', 100)
df= pd.read_csv('sample_text.csv')
df.shape
```

```python
df.head(9)
```

```python
from sentence_transformers import SentenceTransformer

encoder = SentenceTransformer('all-mpnet-base-v2')
vectors= encoder.encode(df.text)
vectors.shape
```

```python
dim= vectors.shape[1]
dim

#every vector has these many dimensions
```

```python
import faiss

index= faiss.IndexFlatL2(dim)
index
```

```python
index.add(vectors)
```

```python
search_query= "I want to buy a Polo t-shirt"

vec= encoder.encode(search_query)
vec.shape
```

```python
vec
```

```python
import numpy as np

svec= np.array(vec).reshape(1,-1)
svec.shape
```

```python
svec
```

```python
index.search(svec, k=2)
```

```python
#first one ie.   array([[1.3799692, 1.3903296]], dtype=float32) ->  distances
#second one ie.  array([[3, 2]])  -> indexes in df

df.head(9) #index 3 and 2 are fashion related like query
```

```python
distances, indexes= index.search(svec, k=2)
indexes
```
